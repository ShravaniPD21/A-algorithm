```python
import heapq

# Tree represented as adjacency list
tree = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3), ('G', 6)],
    'D': [('H', 1), ('I', 4)],
    'E': [],
    'F': [],
    'G': [],
    'H': [],
    'I': []
}

# Heuristic values (estimated cost to goal node 'I')
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 3,
    'E': 6,
    'F': 5,
    'G': 7,
    'H': 3,
    'I': 0
}

def a_star(start, goal):
    open_list = []
    heapq.heappush(open_list, (heuristic[start], start))

    g_cost = {start: 0}
    parent = {start: None}

    print("A* Search Process:\n")

    while open_list:
        f, current = heapq.heappop(open_list)
        print(f"Expanding Node: {current}, f(n) = {f}")

        if current == goal:
            break

        for neighbor, cost in tree[current]:
            new_g = g_cost[current] + cost

            if neighbor not in g_cost or new_g < g_cost[neighbor]:
                g_cost[neighbor] = new_g
                f_cost = new_g + heuristic[neighbor]
                heapq.heappush(open_list, (f_cost, neighbor))
                parent[neighbor] = current

    # Reconstruct path
    path = []
    node = goal
    while node:
        path.append(node)
        node = parent[node]
    path.reverse()

    print("\nFinal Path Found:")
    print(" → ".join(path))
    print("Total Cost:", g_cost[goal])


# Run A* Algorithm
a_star('A', 'I')
```
